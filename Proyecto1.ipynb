```python
import requests
from twilio.rest import Client


credenciales = {}
with open(".env", "r") as f:
    for linea in f:
        clave, valor = linea.strip().split("=")
        credenciales[clave] = valor


API_KEY = credenciales["WEATHER_API_KEY"]
CIUDAD = "Buenos Aires"
URL = f"https://api.openweathermap.org/data/2.5/weather?q={CIUDAD}&appid={API_KEY}&units=metric"


respuesta = requests.get(URL)
datos = respuesta.json()

if respuesta.status_code == 200:
    temperatura = datos["main"]["temp"]
    sensacion_termica = datos["main"]["feels_like"]
    humedad = datos["main"]["humidity"]

    mensaje_clima = f"""
    ğŸŒ Clima en {CIUDAD}:
    ğŸŒ¡ï¸ Temperatura: {temperatura}Â°C
    ğŸ¥µ SensaciÃ³n tÃ©rmica: {sensacion_termica}Â°C
    ğŸ’§ Humedad: {humedad}%
    """

    print(mensaje_clima)


    cliente = Client(credenciales["TWILIO_ACCOUNT_SID"], credenciales["TWILIO_AUTH_TOKEN"])


    mensaje_twilio = cliente.messages.create(
        from_=credenciales["TWILIO_WHATSAPP"],
        body=mensaje_clima,
        to=credenciales["TU_NUMERO"]
    )

    print("âœ… Mensaje enviado con Ã©xito")

else:
    print("âŒ Error al obtener datos del clima. Verifica la API Key o la ciudad.")
```
